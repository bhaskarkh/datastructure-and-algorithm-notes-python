```python
class Node():
    def __init__(self,data=None, next=None):
        self.data=data
        self.next=next
        #print("Node Created of Value= ",data)
class linkedList():
    def __init__(self):
        self.head=None
        
    def insert_at_start(self,data):
        #print("Inserting ",data," at start")
        node=Node(data,self.head)
        self.head=node
        
    def insert_at_end(self,data):
        #print("Inserting ",data," at end")
        if self.head is None:
            self.head=Node(data,None)
            return
        
        itr=self.head
        while itr.next:
            #print("insert at end ===> ",itr.data)
            itr=itr.next
        #print("while finish")
        itr.next=Node(data,None)
    def insert_values(self,givenList):
        self.head=None
        for data in givenList:
            
            self.insert_at_end(data)
    def length(self):
        count=0
        if self.head is not None:
            
            itr=self.head
            while itr.next:
                count+=1
                itr=itr.next
        #print("length= ",count)
        return count
            
    
            
        
        
        
            
    def remove_at(self,index):
        count=0
        if index>0 and index <self.length():
            if self.head is not None:
                itr=self.head
                while itr.next:
                    print("cocunt== ",count," index= ",index-1)
                    if count==index-1:
                        print("Index found removing value = ",itr.data)
                        itr.next=itr.next.next
                        
                        break
                    else:
                        count+=1
                        itr=itr.next
        else:
            print("No element at that index")
    
    def insert_at(self,index,data):
        count=0
        if index==0:
            self.insert_at_start(data)
            return
        else:
            if index>0 and index<self.length():
                itr=self.head
                while itr.next:
                    if count==index-1:
                        # 1--->2--->4--->5--->6--->7--->8
                        #itr.next = 2  itr.next.next=  4
                        print("itr.next =",itr.data," itr.next.next= ",itr.next.data)
                        node=Node(data,itr.next)
                        itr.next=node
                        break
                    else:
                        count+=1
                        itr=itr.next
            else:
                print("provide valid index")
    
    def insert_after_value(self,value,data):
        itr=self.head
        while itr.next:
            if itr.data==value:
                # 1--->2--->4--->5--->6--->7--->8
                #itr.next = 2  itr.next.next=  4
                node=Node(data,itr.next)
                itr.next=node
                break
            else:
                itr=itr.next
    
    def remove_by_value(self,value):
        isFound=False
        if self.head is not None:
            itr=self.head
            while itr.next:
                if itr.next.data==value:
                    print("value found removing ...",itr.next.data)
                    itr.next=itr.next.next

                    break
                else:
                    itr=itr.next
            if isFound:
                print("given element is not there in linked list lindly give element value from below linkedlist")
                self.print()
    
        
        
        
                    
        

            
            
            
        
        
    def print(self):
        itr=self.head
        llstr=''
        if itr is None:
            print("LinkedList is empty")
            return
        
        while itr:
            llstr+= str(itr.data)
            itr=itr.next
            if  itr:
                llstr+="--->"
                
        
        print(llstr)
    
    def insertAtBegining(self,head,x):
        # code here 
        new_node=Node(x)
        new_node.next=head
        return new_node
    
    #Function to insert a node at the end of the linked list.
    def insertAtEnd(self,head,x):
        # code here 
        if head is None:
            head=Node(x)
            return head
        else:
            itr=head
            while itr.next:
                itr=itr.next
            itr.next=Node(x)
            return head
         
            
        
        
```

# Double LinkedList

#### Methods to to be done
1. Insertion
  1.  begin
  1.  end
  1.  mid
2. Deletion
  1.  begin
  1.  end
  1.  mid
3. Traversal
 1. Forward
 1. Backward

```python
class Node():
    def __init__(self,prev=None,data=None,next=None):
        self.prev=prev
        self.data=data
        self.next=next
        
class DoubleLinkedList():
    def __init__(self):
        self.head=None
        
    def insertElementatStart(self,data):
        if self.head is None:
            self.head=Node(None,data,None)
        else:
            self.head=Node(None,data,self.head)
            
    def insertList(self,inputList):
        if len(inputList)==0:
            print("No element present kindly provide list with atleast 2 element")
        else:
            for i in inputList:
                self.insertElementatStart(i)
                
                
    def deleteElementatEndByValue(self,value):
        if self.head is None:
            print("Cannot delete from empty list")
            #90------->>60------->>40------->>20------->>10------->>70
        else:
            itr=self.head
            isFoundFlag=False
            while itr.next:
                if itr.next.data==value:
                    print("Value",value," Found")
                    isFoundFlag=True
                    itr.next=itr.next.next
                    if itr.next.next is not None:
                        itr.next.prev=itr
                        print("element = ",value," deleted")
                        break
                else:
                    itr=itr.next
            if not isFoundFlag:
                print("element not found provide value that is ther in list")
            
                    
                
                
    def print(self):
        ## forward Traversal
        ## from head to last
        if self.head is None:
            print("Empty DoubleLinked List")
        else:
            itr=self.head
            strList=""
            while itr:
                strList+=str(itr.data)
                itr=itr.next
                if itr is not None:
                    strList+="------->>"
            print(strList)
                
                
    
```

```python
dd=DoubleLinkedList()
dd.insertElementatStart(70)
dd.print()
dd.insertList([10,20,40,60,90])
dd.print()
dd.deleteElementatEndByValue(40)
dd.print()
```
